```python
import joblib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from scipy import stats
```

```python
def test(f_loss_file):    
    train_losses = []
    val_losses = []
    val_aurocs = []

    bag_test_aurocs = []

    f_loss = open(f_loss_file)
    count = -1
    params = ''
    for line_loss in f_loss:
        # print(line_loss)
        if count == -1:
            params = line_loss
        count += 1
        if count > 0:
            train_losses.append(float(line_loss.split(', ')[1]))
            val_losses.append(float(line_loss.split(', ')[2]))
            val_aurocs.append(float(line_loss.split(', ')[3]))
            bag_test_aurocs.append(float(line_loss.split(',')[4]))
    # print(len(val_aurocs) - np.argmax(val_aurocs))
    return bag_test_aurocs[np.argmax(val_aurocs)]
```

```python


gpi = [100.0, 88.8888888888889, 77.77777777777779, 66.66666666666666, 55.55555555555556, 44.44444444444445, 33.333333333333336, 22.222222222222225, 11.111111111111112, 0.0]
all_models = ['abdmil_pe', 'transmil', 'DAS_transformer']


COUNT = -1
labs = ['a', 'b', 'c']
for num_bags in [100]:
    bagrocs = {}
    bagrocs_iqr = {}
    fig, ax = plt.subplots()
    fig.set_figheight(4.5)
    fig.set_figwidth(8)
    COUNT += 1
    for model in all_models:
        bagrocs[model] = []
        bagrocs_iqr[model] = []
        

    for perm_dist in range(1, 11):
        for model in all_models:
            currbr = []
            for seed in range(10):
                
                try:
                    f_loss_file = '/data2/meerak/txtfiles/seed%d_numbags%d_MNIST_permdist%d_%s_loss.txt'%(seed, num_bags, perm_dist, model)
                    currbr.append(test(f_loss_file))
                except:
                    error = 1
                    currbr.append(0)

            print(perm_dist, gpi[perm_dist - 1], model)
            print(currbr)
            print('%0.3f (%0.3f, %0.3f)'%(np.median(currbr), np.median(currbr) - stats.iqr(currbr)/2, np.median(currbr) + stats.iqr(currbr)/2))
            
            bagrocs[model].append(np.median(currbr))
            bagrocs_iqr[model].append(stats.iqr(currbr)/2)
            
            print()
        print()

    colors = ['red', 'green', 'purple']
    for i, model in enumerate(['abdmil_pe', 'transmil', 'DAS_transformer']):
        ax.scatter(gpi, np.array(bagrocs[model]), color = colors[i])
        ax.plot(gpi, np.array(bagrocs[model]), color = colors[i])
        ax.fill_between(gpi, np.array(bagrocs[model]) - np.array(bagrocs_iqr[model]), np.array(bagrocs[model]) + np.array(bagrocs_iqr[model]), alpha=0.2, color = colors[i])

    for tick in ax.xaxis.get_major_ticks():
        tick.label.set_fontsize(15)
    for tick in ax.yaxis.get_major_ticks():
        tick.label.set_fontsize(15)
    ymin, ymax = ax.get_ylim()
    xmin, xmax = ax.get_xlim()
    ax.set_xticks([0, 33, 66, 100], labels = [0, 33, 66, 100])

    ax.set_ylabel('AUROC', fontsize=16)
    ax.set_xlabel('% Absolute Position Information', fontsize=18)



    fig.tight_layout()
    fig.savefig('das.png')

```
